### Feature / Target Split

```python
X = df.drop([ c for c in df.columns if c.startswith('product_')], axis=1).drop(['CaseId'], axis=1)
y = df[[ c for c in df.columns if c.startswith('product_')]].values
```

```python
y_clean = []
# remove columns with too few entries for target
for i in range(y.shape[1]):
    if y[:,i].sum() >= 2:
        y_clean.append(y[:,i])
y_clean = np.transpose(np.array(y_clean))
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_clean, test_size=0.1, random_state=42)
```

```python
X_train_base_group = X_train[[c for c in X_train.columns if c.startswith('base_group_')]]
#base_group_pca = PCA(n_components=50)
X_test_base_group = X_test[[c for c in X_train.columns if c.startswith('base_group_')]]

X_train_base_id = X_train[[c for c in X_train.columns if c.startswith('base_id_')]]
X_test_base_id = X_test[[c for c in X_train.columns if c.startswith('base_id_')]]
base_id_pca = PCA(n_components=100)
X_train_base_id_tr = base_id_pca.fit_transform(X_train_base_id)
X_test_base_id_tr = base_id_pca.fit_transform(X_test_base_id)

X_train_symptom = X_train[[c for c in X_train.columns if c.startswith('symptom_')]]
X_test_symptom = X_test[[c for c in X_train.columns if c.startswith('symptom_')]]
```

```python
X_train = np.concatenate([X_train_base_group.values, X_train_base_id_tr, X_train_symptom.values], axis=1)
X_test = np.concatenate([X_test_base_group.values, X_test_base_id_tr, X_test_symptom.values], axis=1)
```

## Train

### SGD Classifier

```python
clf = MultiOutputClassifier(SGDClassifier(max_iter=1000, tol=1e-3)) # n_jobs
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
zero_one_loss(y_test, y_pred)
```

```python
hamming_loss(y_test,y_pred)
```

### XGBoost

```python
from xgboost import XGBClassifier
```

```python
clf = MultiOutputClassifier(XGBClassifier(n_jobs=-1, max_depth=4))
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
zero_one_loss(y_test, y_pred)
```

```python
hamming_loss(y_test,y_pred)
```

### Keras

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_clean, test_size=0.1, random_state=42)
```

```python
y_test.shape
```

```python
# pip install tensorflow
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense
```

```python
inputs = Input(shape=(X_train.shape[1],))

# x = Dense(1024, activation='relu')(inputs)
x = Dense(512, activation='relu')(inputs)
x = Dense(256, activation='relu')(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)

outputs = Dense(y_train.shape[1], activation='sigmoid')(x)
```

```python
model = keras.Model(inputs=inputs, outputs=outputs)
model.summary()
```

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='rmsprop',
    metrics=["accuracy"],
)

history = model.fit(X_train, y_train, batch_size=64, epochs=100)
```

```python
y_pred = model.predict(X_test).round()
```

```python
zero_one_loss(y_test, y_pred)
```

```python
hamming_loss(y_test,y_pred)
```
